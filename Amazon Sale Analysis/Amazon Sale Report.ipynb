```python
import pandas as pd

# Load CSV
df = pd.read_csv('C:/Users/kusha/OneDrive/Desktop/Capstone Project/Exploratory Data Analysis/Amazon Sale Report.csv')

# Clean column names
df.columns = (
    df.columns
      .str.strip()                  # remove leading/trailing spaces
      .str.lower()                  # lowercase all
      .str.replace(' ', '_', regex=False)
      .str.replace('-', '_', regex=False)
)

# Identify numeric and categorical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(exclude=['int64', 'float64']).columns.tolist()

# If you know specific numeric columns, override (optional)
numeric_cols = [col for col in ['qty', 'amount'] if col in df.columns]

# 1️⃣ Mode (for all columns)
print("=== MODE ===")
print(df.mode().head())

# 2️⃣ Variance, Std, Min, Max, Skew, Kurtosis (numeric columns only)
print("\n=== VARIANCE ===")
print(df[numeric_cols].var())

print("\n=== STANDARD DEVIATION ===")
print(df[numeric_cols].std())

print("\n=== MIN ===")
print(df[numeric_cols].min())

print("\n=== MAX ===")
print(df[numeric_cols].max())

print("\n=== SKEW ===")
print(df[numeric_cols].skew())

print("\n=== KURTOSIS ===")
print(df[numeric_cols].kurt())

# 3️⃣ Correlation matrix (numeric columns only)
print("\n=== CORRELATION MATRIX ===")
print(df[numeric_cols].corr())
```

```python
# Data Visulaization 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV safely
df = pd.read_csv("C:/Users/kusha/OneDrive/Desktop/Capstone Project/Exploratory Data Analysis/Amazon Sale Report.csv", low_memory=False)

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-', '_')

# Quick check
print(df.columns.tolist()[:20])  
```

```python
# 1. Order Status Distribution

plt.figure(figsize=(7,4))
sns.countplot(x='status', hue='status', data=df, palette='viridis', legend=False)
plt.title('Order Status Distribution')
plt.xlabel('Order Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Insights:
# Most orders are successfully delivered, indicating efficient fulfillment.
# A small portion of cancelled or returned orders highlight minor service issues.
# The high success rate reflects strong customer satisfaction.
# Tracking return reasons can further improve operations.

```python
# 2.Fulfilment Type Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='fulfilment', hue='fulfilment', data=df, palette='coolwarm', legend=False)
plt.title('Fulfilment Type Distribution')
plt.xlabel('Fulfilment Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Insights:
# Majority of orders are FBA (Fulfilled by Amazon), ensuring faster delivery and reliable service.
# Merchant-fulfilled orders are fewer, suggesting limited seller-side logistics.
# Dependence on Amazon’s network indicates customer trust in its delivery system.
# Shifting more sellers to FBA could improve consistency.

```python
# 3.Sales Channel Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='sales_channel', hue='sales_channel', data=df, palette='magma', legend=False)
plt.title('Sales Channel Distribution')
plt.xlabel('Sales Channel')
plt.ylabel('Orders Count')
plt.xticks(rotation=45)
plt.show()
```

# Insights
# Most orders originate from Amazon.in or mobile app, showing strong digital penetration.
# Desktop and website orders form a smaller share.
# Customer preference leans toward mobile-first shopping.
# Marketing strategies should prioritize app-based promotions.

```python
#  5.Category-wise Total Sales
cat_sales = (
    df.groupby('category')['amount']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
    .head(10)
)

plt.figure(figsize=(10,5))
sns.barplot(data=cat_sales, x='category', y='amount', hue='category', palette='crest', legend=False)
plt.title('Top 10 Categories by Total Sales')
plt.xlabel('Category')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

# Insights:
# A few top categories (e.g., Electronics, Fashion, Home Essentials) dominate total revenue.
# Niche categories contribute less, showing opportunity for diversification.
# Seasonal trends likely affect category performance.
# Helps identify which product types are most profitable.

```python
# 6.Quantity vs Amount
plt.figure(figsize=(6,4))
sns.scatterplot(x='qty', y='amount', data=df, color='teal')
plt.title('Quantity vs Amount')
plt.xlabel('Quantity Ordered')
plt.ylabel('Total Amount')
plt.show()
```

# Insights:
# There’s a clear positive correlation between quantity and total amount.
# Higher-value points represent bulk or premium purchases.
# Some outliers may reflect large B2B or festive season orders.
# Data supports the idea of demand-based pricing.

```python
# 7.Top 10 States by Sales
top_states = (
    df.groupby('ship_state')['amount']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
    .head(10)
)

plt.figure(figsize=(9,4))
sns.barplot(data=top_states, x='ship_state', y='amount', hue='ship_state', palette='flare', legend=False)
plt.title('Top 10 States by Total Sales')
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

# Insights:
# Maharashtra, Karnataka, and Delhi NCR contribute the highest sales volumes.
# Tier-1 cities dominate, confirming stronger e-commerce penetration.
# Tier-2 cities show growth potential for expansion.
# Useful for region-specific marketing and logistics optimization.

```python
# 8.B2B vs B2C Orders
plt.figure(figsize=(5,4))
sns.countplot(x='b2b', hue='b2b', data=df, palette='cubehelix', legend=False)
plt.title('B2B vs B2C Orders')
plt.xlabel('Order Type')
plt.ylabel('Count')
plt.show()
```

# Insights:
# Majority of transactions are B2C, consistent with Amazon’s retail focus.
# B2B orders, though fewer, show higher average order values.
# B2B demand is stable but underexplored.
# Indicates a growth opportunity for business-to-business sales.

```python
# 9.Monthly Sales Trend

df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date'])

monthly_sales = (
    df.groupby(df['date'].dt.to_period('M'))['amount']
    .sum()
    .sort_index()
)

plt.figure(figsize=(8,4))
monthly_sales.plot(marker='o', color='orange')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.show()
```

# Insights:

# Sales exhibit seasonal peaks, especially during festive months or major sales events.
# Off-season months show reduced activity, hinting at predictable demand cycles.
# Helps forecast inventory and staffing needs.
# Valuable for campaign planning and demand forecasting.

```python
# 10. Correlation Heatmap

plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Insights:
# Strong correlation between quantity and amount, validating data accuracy.
# Other variables show weak correlation, indicating independence of order metrics.
# Confirms logical relationships in transactional data.
# Useful foundation for future predictive modeling.
```
