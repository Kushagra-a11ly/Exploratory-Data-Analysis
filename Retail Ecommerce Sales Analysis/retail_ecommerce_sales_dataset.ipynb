```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set(style="whitegrid")

# 1. Sales Distribution
plt.figure(figsize=(7,4))
plt.hist(df['Sales'], bins=30)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

Sales Distribution Insights

    1.Right-skewed distribution: Peak frequency (~175) occurs at $1,000-$2,000, with most sales concentrated in lower value ranges
    2.High concentration at low values: Approximately 70-80% of all sales fall below $4,000, indicating the business is primarily driven by smaller transactions
    3.Steep initial decline: Frequency drops dramatically from the peak to $4,000, suggesting a natural price sensitivity threshold
    4.Extended tail: Sales continue to ~$13,000 with very low frequency (<10), representing rare premium transactions or bulk orders
    5.Three customer segments emerge: Core customers ($0-$4,000), mid-tier ($4,000-$8,000), and premium accounts ($8,000+)
    6.Strategic recommendation: Optimize operations for high-volume, lower-value transactions while providing dedicated attention to rare but valuable premium customers

```python
# 2. Profit Distribution (Boxplot)
plt.figure(figsize=(7,4))
sns.boxplot(x=df['Profit'])
plt.title("Profit Distribution")
plt.show()
```

# Insights 

    # 1.Profit values are highly right-skewed, with most profits concentrated on the lower side.
    # 2.A large number of outliers are present, mainly between 1500–3500, indicating rare but very high-profit transactions.
    # 3.The IQR (middle 50%) is narrow, roughly between 250–750, showing that most orders generate moderate profits.
    # 4.The median profit lies in the lower region, confirming that typical profits are relatively small.
    # 5.The upper whisker is long, reflecting significant variability among higher-profit transactions.
    # 6.The lower whisker is short, indicating consistent low-profit values with less variation.

```python
# 3. Category Count
plt.figure(figsize=(7,4))
sns.countplot(x=df['Category'])
plt.title("Order Count by Category")
plt.show()
```

# Insights

    # 1.Technology has the highest number of orders, slightly above the other categories.
    # 2.Furniture has a similar order volume to Technology, indicating balanced demand.
    # 3.Office Supplies has the lowest order count, though the difference is not very large.
    # 4.Order counts across all three categories are fairly consistent, showing no extreme imbalance.
    # 5.The chart suggests customers purchase items from all categories almost equally, indicating diverse product demand.
    # 6.No category shows exceptionally high or low demand, which reflects a well-distributed sales pattern across product types.

```python
# 4. Sales vs Profit (Scatterplot)
plt.figure(figsize=(7,4)) 
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Sales vs Profit")
plt.show()
```

# Insights
    # 1.Positive Correlation – As sales increase, profit generally increases as well, indicating a strong positive relationship between the two variables.
    # 2.Wider Spread at Higher Sales – At lower sales values, profit values are tightly clustered, but as sales rise, profit values become more widely spread. This suggests greater variability and risk at higher sales levels.
    # 3.Most Data in Mid-Range – A large concentration of points lies roughly between 2000–7000 in sales and 200–1500 in profit, indicating this is the most common business performance range.
    # 4.Some High-Sales Low-Profit Cases – A few points show high sales but relatively low profit, which might indicate inefficiencies, high costs, or discount-driven sales.
    # 5.Possible Outliers – Points in the far upper-right (e.g., sales above 10,000 and profit above 3000) appear as outliers, representing unusually high-performing cases.
    # 6.Nonlinear Profit Growth – Profit does not increase perfectly linearly with sales; the scatter suggests diminishing returns or varying margins at higher sales levels.

    # 6.Selling more units generally increases both revenue and profit, but not as strongly as price does.

```python
# 7. Total Sales by Category
plt.figure(figsize=(7,4))
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Total Sales by Category")
plt.ylabel("Sales")
plt.show()
```

```python
# 8. Monthly Sales Trend
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Month'] = df['Order_Date'].dt.to_period('M').astype(str)
monthly_sales = df.groupby('Month')['Sales'].sum()
plt.figure(figsize=(10,4))
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

# Insights
    # 1.Sales show frequent ups and downs, indicating a highly fluctuating monthly trend.
    # 2.The highest spike occurs around late 2022, crossing roughly 280,000.
    # 3.Early 2022 starts strong but quickly dips, showing inconsistent demand.
    # 4.Mid-2022 sees noticeable volatility with sharp rises and drops month to month.
    # 5.Early 2023 includes another strong surge before declining again.
    # 6.Late 2023 stabilizes somewhat, but still ends with significant variation.

```python
# 9. Top 10 Products by Sales
top10 = df.groupby('Product_ID')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(7,4))
top10.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

# Insights
    # 1.PROD-674 leads all products with the highest total sales, clearly outperforming the rest.
    # 2.The next two strong performers, PROD-401 and PROD-232, also show consistently high sales close to each other.
    # 3.There is a gradual decline in sales as we move from the top product toward the lower-ranked items.
    # 4.The sales difference between mid-ranked products (PROD-180, PROD-188, PROD-988) is relatively small, indicating similar demand.
    # 5.Products ranked 7 to 10 (PROD-949, PROD-319, PROD-682, PROD-409) still maintain strong sales, suggesting a healthy long-tail performance.
    # 6.Overall, the top 10 products have fairly tight sales ranges, showing no extreme outlier except the leading product.

```python
from scipy.stats import zscore
z = zscore(df['Sales'])
print(z[:10]) 
```

```python
# IQR Method
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['Sales'] < Q1 - 1.5*IQR) | (df['Sales'] > Q3 + 1.5*IQR)]

print(outliers.head())       
print(len(outliers))
```

```python
#  Outlier Percentage

percentage = (len(outliers) / len(df)) * 100
print(f"Outlier Percentage: {percentage:.2f}%")
```
